```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
ax = sns.histplot(data=df,x='fare',bins=30)
ax.set(xlabel='fare',ylabel='Frequency',title='fare distribution with hist')
plt.show()

ax = sns.histplot(data=df,x='age',bins=30)
ax.set(xlabel='age',ylabel='Frequency',title='age distribution with hist')
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
sns.countplot(data=df,x='survived')
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
sns.countplot(data=df, x='pclass')
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
g = sns.JointGrid(data=df,x='fare',y='age')
g.plot_joint(sns.scatterplot, edgecolor = None)
g.ax_marg_x.hist(df['fare'], color = (0,0,.8,.1), edgecolor = (0,0,.8,.5), bins = 40)
g.ax_marg_y.hist(df['age'], orientation = 'horizontal', color = (0,0,.8,0.1), edgecolor = (0,0,.8,0.5), bins = 30)
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
ax = sns.displot(df['fare'],bins=30,kde=False,color='salmon')
ax.set(ylabel=None)
plt.margins(x=0,y=0)
plt.xlim(0, 600)
plt.ylim(0,500)
sns.despine(right=False)
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
sns.boxplot(data=df,x='class',y='age',palette='pastel')
plt.ylim(0,80)
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
ax = sns.swarmplot(data=df,x='class',y='age',size=5)
ax.set(yticks=[0,20,40,60,80])
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
sns.countplot(data=df,x='sex')
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!

titanic_mx = df.corr()
ax = sns.heatmap(titanic_mx, annot=False, cmap='coolwarm')
ax.set(title='titanic.corr()')
ax.tick_params(axis='x',rotation=0)
plt.show()
```
