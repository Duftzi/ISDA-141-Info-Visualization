Q1 Generate the following plots. I have done stacked plot

```python
# stacked bar chart
# some simple data 
x = [1,2,3,4]
y = [5,4,3,2]
y1 = [7,8,5,3]

# create stacked bar charts 
plt.subplot(234)
plt.bar(x, y)


# we need more data for stacked bar charts 

plt.bar(x, y1, bottom=y, color = 'r')

plt.title('stacked bar chart')

# line chart
# some simple data
x = [1,2,3,4]
y = [5,4,3,2]

# create line chart
plt.subplot(231)
plt.plot(x,y)

# more data

plt.xticks([2,4])
plt.title('line chart')

# bar chart
# some simple data
x = [1,2,3,4]
y = [5,4,3,2]

# create bar chart
plt.subplot(232)
plt.bar(x,y)

# more data

plt.title('Bar chart')

# horizontal bar chart
# some simple data
x = [1,2,3,4]
y = [5,4,3,2]

# create horizontal bar chart
plt.subplot(233)
plt.barh(x,y)

# more data

plt.yticks([2,4])
plt.title('Horizontal Bar Chart')

# box plot
# some simple data
x = [1,2,3,4]

# create box plot
plt.subplot(235)
plt.boxplot(x)

# more data

plt.title('Box Plot')

# scatter plot
# some simple data
x = [1,2,3,4]
y = [5,4,3,2]

# create scatter plot
plt.subplot(236)
plt.scatter(x,y)

# more data

plt.xticks([2,4])
plt.title('Scatter Plot')

plt.tight_layout()
plt.show()
```

Q4. Create the following bar chart with error bar 

```python
x = ['Nuclear', 'Hydro', 'Gas', 'Oil', 'Coal', 'Biofuel']
energy = [50, 60, 150, 220, 240, 200]
variance = [10, 20, 20, 15, 20, 30]
plt.errorbar(x, energy, color='green', yerr=variance)
plt.xlabel("Energy Source")
plt.ylabel("Energy Output")
plt.title("Energy output from various fuel sources")
```

Q5 Create a following stacked bar chart

```python
countries = ['USA', 'GB', 'China', 'Russia', 'Germany']
bronzes = np.array([38, 17, 26, 19, 15])
silvers = np.array([37, 23, 18, 18, 10])
golds = np.array([46, 27, 26, 19, 17])

# create stacked bar charts
plt.bar(countries, golds, bottom = silvers+bronzes, color = '#FFD700')
plt.bar(countries, silvers, bottom = bronzes, color = '#C0C0C0')
plt.bar(countries, bronzes, color = '#CD7F32')

# we need more data for stacked bar charts 

plt.title('2012 Olympics Top Scorers')
plt.legend(['golds','silvers','bronzes'])
plt.xlabel('Countries')
plt.ylabel('Medals')

plt.tight_layout()
```

```python
# create pie chart
Tasks = [200,300,600]

my_labels = 'C1','C2','C2'
my_colors = ['blue','red','green']

# my code
explode = (0,0.1,0) # explode C2
plt.pie(Tasks, explode=explode,labels=my_labels, colors=my_colors, autopct='%.1f%%', shadow=True, startangle=15)

# more data
plt.title('My Tasks')

# show plot
plt.show()
```
