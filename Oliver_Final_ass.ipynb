```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_titanic = pd.read_csv("titanic.csv")
df_titanic.head()
```

Q3 Consider the following features and draw some conclusions about survival or no survival using any kind of visualization.

    1. age
    2. Pclass
    3. Sex
      
- Fare and age


```python
## Your code here
## Q3: Based on the graphs we can see that many people aged between 20-40 died on the titanic. 
## Many people in Pclass 3 did not survive the titanic. More men died than women in the titanic.
## It also seems that the fare does not have much to do with the deaths on the titanic. 
titanicAge = sns.FacetGrid(df_titanic, col='Survived')
titanicAge.map(plt.hist, 'Age', bins=20)

titanicPclass = sns.FacetGrid(df_titanic, col='Survived')
titanicPclass.map(plt.hist, 'Pclass', bins=20)

titanicSex = sns.FacetGrid(df_titanic, col='Survived')
titanicSex.map(plt.hist, 'Sex', bins=20)

titanicFare = sns.FacetGrid(df_titanic, col='Survived')
titanicFare.map(plt.hist, 'Fare', bins=20)
```

Q4. Make the pair plot of the following features. Color your plot using the "Sex" feature.

```python
## Your code here
## From the graph we can see that middle aged families had more children on the titanic. 
## We can also see from the graph that people aged 0-25 were more likely to have siblings. 
## There were much more males on the titanic given by data in the pclass. 
## People of all ages were on the titanic when it sank. 
## A lot of people did not have children/spouses on the titanic.
sns.pairplot(df_titanic,x_vars=["Survived", "Pclass", "Age", "Siblings/Spouses Aboard", "Parents/Children Aboard"],
             y_vars=["Survived", "Pclass", "Age", "Siblings/Spouses Aboard", "Parents/Children Aboard"],hue="Sex", corner=True)
```
